```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
path='ml-25m/'
rating_data_file = path +'/ratings.csv'
movie_data_file=path+'movies.csv'
```

```python
movie_df = pd.read_csv(rating_data_file)
```

```python
movie_df.head()
```

```python
movie_titles = pd.read_csv(movie_data_file)
movie_titles.head()
```

```python
movie_df = pd.merge(movie_df,movie_titles, on='movieId')
movie_df.head()
```

```python
movie_df.describe()
```

```python
movie_df.dropna(inplace=True)
movie_df.shape
```

```python
ratings = pd.DataFrame(movie_df.groupby('title')['rating'].mean())
```

```python
ratings['number_of_ratings'] = movie_df.groupby('title')['rating'].count()
ratings.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
ratings['rating'].hist(bins=50)
plt.show();
```

```python
ratings['number_of_ratings'].hist(bins=60)
plt.show();
```

```python
import seaborn as sns
sns.jointplot(x='rating',y='number_of_ratings',data=ratings)
plt.show();
```

```python
ratings.sort_values(by=['number_of_ratings'], ascending=False, inplace=True)
```

```python
ratings.head()
```

```python
top_rating = ratings[0:500]
```

```python
top_rating
```

```python
movie_df.head()
```

```python
top_rating.reset_index(inplace=True)
```

```python
top_rating.head()
```

```python
topmovies_df = movie_df[movie_df['title'].isin(top_rating['title'])]
```

```python
topmovies_df.shape
```

```python
movie_df.shape
```

```python
movie_matrix = topmovies_df.pivot_table(index='userId', columns='title',
                                        values='rating')
```

```python
movie_matrix.head()
```

```python
top_rating.sort_values('number_of_ratings', ascending=False).head(10)
```

```python
FG_user_rating = movie_matrix['Forrest Gump (1994)']
```

```python
FG_user_rating.head()
```

```python
similar_to_fg = movie_matrix.corrwith(FG_user_rating)
```

```python
similar_to_fg.sort_values(ascending=False).head()
```

```python
SL_user_rating = movie_matrix['Schindler\'s List (1993)']
```

```python
SL_user_rating.head()
```

```python
similar_to_sl = movie_matrix.corrwith(SL_user_rating)
```

```python
similar_to_sl.sort_values(ascending=False).head()
```

```python
corr_SL = pd.DataFrame(similar_to_sl, columns=['correlation'])
corr_SL.dropna(inplace=True)
corr_SL.head()
```

```python
corr_SL.shape
```

```python
corr_FG = pd.DataFrame(similar_to_fg, columns=['correlation'])
corr_FG.dropna(inplace=True)
corr_FG.head()
```

```python
top_rating.set_index(['title'], inplace=True)
top_rating.head()
```

```python
corr_SL = corr_SL.join(top_rating).sort_values(['correlation'], ascending=False)
corr_SL.head()
```

```python
corr_FG = corr_FG.join(top_rating).sort_values(['correlation'], ascending = False)
corr_FG.head()
```
